```python
def hdd(a):
    if "HDD" in a:
        if "+" in a:
            a=a.split("+")
            total=0
            for i in a:
                if "HDD" in i:
                    i=i.replace("HDD","")
                    if "GB" in i:
                        i=i.replace("GB","")
                        total=total+float(i)
                    elif "TB" in i:
                        i=i.replace("TB","")
                        i=float(i)*1000
                        total=total+i
            return total
        elif "GB" in a:
            a=a.replace("HDD","")
            a=a.replace("GB","")
            return int(a)
        elif "TB" in a:
            a=a.replace("HDD","")
            a=a.replace("TB","")
            return float(a)*1000
    else:
        return 0
                            
```

```python
def ssd(a):
    if "SSD" in a:
        if "+" in a:
            a=a.split("+")
            total=0
            for i in a:
                if "SSD" in i:
                    i=i.replace("SSD","")
                    if "GB" in i:
                        i=i.replace("GB","")
                        total=total+float(i)
                    elif "TB" in i:
                        i=i.replace("TB","")
                        i=float(i)*1000
                        total=total+i
            return total
        elif "GB" in a:
            a=a.replace("SSD","")
            a=a.replace("GB","")
            return int(a)
        elif "TB" in a:
            a=a.replace("SSD","")
            a=a.replace("TB","")
            return float(a)*1000
    else:
        return 0
                            
```

### Splitting the data

```python
x=df.drop("Price",axis=1)
y=np.log(df["Price"])
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
x.iloc[:,:]=sd.fit_transform(x.iloc[:,:])
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

### 1 - Predicting the data using Linear egression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
```

```python
from sklearn.metrics import r2_score
r2_score(ytest,ypred)
```

```python
plt.scatter(ytest,ypred)
plt.show()
```

```python
sns.distplot(ytest-ypred)
plt.show()
```

### we have achieved an average accuracy of 75 % which isnt that great.

```python
lr.score(xtrain,ytrain)
```

```python
lr.score(xtest,ytest)
```

### Evaluate the model(Errors)

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
print("MAE:",mean_absolute_error(ytest,ypred))
print("MSE:",mean_squared_error(ytest,ypred))
print("RMSE:",np.sqrt(mean_squared_error(ytest,ypred)))
```

### From the above values, we can conclude that the model is not over fitting. However it a very low accuracy
